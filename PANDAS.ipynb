```python
# CREATING DATAFRAME 1

import pandas as pd
data1={
    "ID":["1","2","3","4","5"],
    "Feature1":["A","B","C","D","E"],
    "Feature2":["L","M","N","O","P"]
}
df1=pd.DataFrame(data1,columns=["ID","Feature1","Feature2"])
df1
```

```python
# CREATING DATA FRAME 2

import pandas as pd
data2={
    "ID":["1","3","6","7","8"],
    "Feature1":["AB","BC","CD","DE","EF"],
    "Feature2":["KL","LM","M-N","N-O","O-P"]
}
df2=pd.DataFrame(data2,columns=["ID","Feature1","Feature2"])
df2
```

```python
# MERGING

# 1.. OUTER

DF_merge_o=pd.merge(df1,df2, on="ID",how="outer")
DF_merge_o  # it includes all the data but the common reference column names common values will be present but for remaining ID'S it will be NaN


# 2.. INNER

DF_merge_i=pd.merge(df1,df2, on="ID",how="inner")
DF_merge_i  # it takes the common values from both the dataframes means interection point

#3.. RIGHT
DF_merge_r=pd.merge(df1,df2, on="ID",how="right")
DF_merge_r  # it takes the inpersection points and right values


# 4.. LEFT
DF_merge_l=pd.merge(df1,df2, on="ID",how="left")
DF_merge_l #  it takes the inpersection points and left values


# 5.. CROSS
DF_merge_c=pd.merge(df1,df2, on="ID",how="left")
DF_merge_c

#  DIFFERENCE B/W CONCAT,JOIN,MERGE

# CONCAT.. SAME COLUMN NAMES
# JOIN.... DIFFERENT COLUMN NAMES
# MERGE... REFERENCE COLUMN NAME MUST BE SAME.. OTHER COLUMN NAMES MAY BE SAME OR DIFFER
```
